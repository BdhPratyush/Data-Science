### Imporing Data


```python
import pandas as pd 
df = pd.read_csv('Health_heart_experimental.csv')
df.head(15)
```

### Line Plot

```python

sns.lineplot(x='age', y='SysBP', data=df)
```

In the above line graph, we can see people with higher age precisely from 50 and onwards, their systolic blood pressure is in the higher side. This indicates that with the increase in age, there is some increase in the heart rate. The data indicates that their blood pressure is quite in the higher side. The maximum blood pressure as seen from the above graph is almost 126.5 which should be considered as it might be an indicate for high blood pressure and its consequences.

### Histogram


```python
plt.hist(df['BMI'], bins=10, color='skyblue', edgecolor='black')
```

The histogram shows the distribution of Body Mass Index(BMI) values, where the x-axis represents the BMI value and the y-axis represents the number of people. The distribution is quite left skewed, with most observations of people at higher BMI values with lower number of people having lower BMI values. The majority of individuals fall within the BMI range of 25-40, indicating a high presence of overweight and obese people, while in comparison very few people with low BMI values are present. This histogram suggests that most people fall under high BMI values, which indicate they are prone to various health risks, which should be controlled in time.

### Bar Plot

```python
sns.barplot(x='sex', y='BMI', data=df)
```

The bar chart compares the mean Body Mass Index across the two sex categories, coded as 0 and 1. The results show that the average BMI for both the sex groups is same, at around 28-29, with almost no noticeable difference between them. By observing the graph, we can say that within this dataset, the BMI is independent of sex. That means, the same alignment of bar heights is not a distinguishing factor for BMI when considering average values.

### Box Plot


```python
sns.boxplot(df['HR'])
```

The above box plot shows the distribution of Heart Rates of people. It is observed that the median value for heart rate is approximately 140 BPM, which is quite high, so it can be interpreted as the data is taken right after some exercises or some physical acitivities. The range of the data is from 80 BPM to 200 BPM. Heart rate of around 140-150 BPM is considered healthy after a heavy physical workout, while having the heart beat at higher frequency than that could lead to a serious health hazard. Most of our dataset lies between 110-170 BPM which can be considered healthy, or just inside the healthy-range.

### Pie Chart

```python
df['indication'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

The pie chart shows the BMI indication where 0 indicates healthy BMI whereas unhealthy BMI is indicated by 1. Here, it is shown that 52.7% of individuals have a healthy BMI(0), while 47.3% have an unhealthy BMI (1). In general, 18-25 BMI is considered healthy, and above it is considered unhealthy.The distribution is quite balanced, half the people are very healthy whereas the other half is quite in risk of health problems. It highlights the importance of monitoritng BMI, as almost half of the people fall outside the healthy range, which is quite concerning. The large percentage of unhealthy BMI range is due to the reckless eating habits, physical dullness, which is quite often seen.

### Scatter Plot

```python
healthy = df[df['indication'] == 0]
unhealthy = df[df['indication'] == 1]

plt.figure(figsize=(8,6))

plt.scatter(healthy['heightCm'], healthy['weightKg'], alpha=0.7, label='Healthy BMI')
plt.scatter(unhealthy['heightCm'], unhealthy['weightKg'], alpha=0.7, label='Unhealthy BMI')

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Weight vs Height by BMI Status')
plt.legend()

# Automatically adjust spacing
plt.tight_layout()

plt.show()
```

The weight versus height scatterplot shows distinct diagonal bands of scatter due to BMI being a derived variable from height and weight. A triangular distribution is produced by the division of healthy and unhealthy BMI groups according to constant BMI boundaries. The plot is mathematically accurate, but rather than being a distinct graph to be examined and a characteristic to be noted, it mainly represents the BMI computation.
