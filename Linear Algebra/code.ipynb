### Basic Matrix Operations

1. Matrix Operation Perform the following operations on the given matrices: A=[[2 -1], [​0 3​]] B=[[1 ​ 2], [-3 4​]] a. A+B b. AB c. 3A−2B

```python
from matplotlib.pylab import beta
import numpy as np
import numpy.linalg as la
```

Define two matrices A and B

```python
A = np.array([[2, -1], [0, 3]])
B = np.array([[1, 2], [3, 4]])
print("Matrix A:\n", A)
print("\n Matrix B:\n", B)
```

##### Matrix Addition

```python
C = A + B
print("\n Matrix Addition:\n A + B =", C)
```

#### Matrix Multiplicaiton

```python
D = np.dot(A, B)
# We can also use D=A@B and A.dot(B)
print("\n Matrix Multiplication:\n A * B =", D)
```

#### Matrix Subtraction

```python
E = 3 * A - 2 * B
print("\n Matrix Subtraction:\n 3A - 2B =", E)
```

### Determinant and Inverse

2. Find the determinant of A. Determine whether A is invertible. If yes, calculate A^−1.

#### Determinant of matrix A

```python
print("Matrix A=\n", A)
det_A = la.det(A)
print("\n Determinant of A:", det_A)
```

#### Inverse of A

```python
if det_A != 0:
    inv_A = la.inv(A)
    print("\n Inverse of A:\n", inv_A)
```

### Eigen Values and Eigen Vectors

3. Compute the eigenvalues and eigenvectors of the matrix:
C=[[4    1], [​−2       1]​]


```python
C = np.array([[4, 1], [-2, 1]])
eigenvalues, eigenvectors = la.eig(C)

# Print the results
print("\n Matrix C:")
print(C)
print("\n Eigenvalue (first): ")
print(eigenvalues[0])
print("\n Corresponding Eigenvector:")
print(eigenvectors[:, 0])
```

4. You are given a dataset of cars with the following features:Number of cylinders (x1​)Engine size in liters (x2​)Horsepower (x3)
car 1: x1: 4 x2:  1.5 x3: 100 and price: 20

    car 2: x1: 6 x2:  2.0 x3: 150 and price: 30

    Then, find the price of a car 3 with features
    x1: 8 x2:  3.0 x3: 200 price:?

```python
X = np.array([
    [4, 1.5, 100],   
     [6, 2.0, 150]   
])

Y = np.array([20, 30])


beta = np.linalg.pinv(X.T @ X) @ X.T @ Y

# beta = np.linalg.pinv(X) @ Y

# beta = np.linalg.lstsq(X, Y,rcond=None)[0] #<- returns a tuple with first elements solution , then other stuffs too
print("β =", beta)

car3= np.array([
    [8, 3.0, 200]
])

price_car3 = car3 @ beta

#dot product returns a matrix so [0]
print(f" price of car 3 : {price_car3[0]}")
```
